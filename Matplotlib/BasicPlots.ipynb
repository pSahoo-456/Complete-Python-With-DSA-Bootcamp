```python
import matplotlib.pyplot  as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#Create a line plot

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axix')
plt.title('Basic Line Plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#create a coustomozed plot
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
#Bar plot
catagories=['A','B','C','D','E']
values=[5,7,3,1,8]

#Create a bar plot
plt.bar(catagories,values,color='purple')
plt.xlabel('Catagories')
plt.ylabel('Values')
plt.show()
```

```python
#scatter plot
x=[1,2,3,4,5]
y=[2,4,5,8,9]
plt.scatter(x,y,color='green',marker='*')
```

```python
#piechart

labels=['A','B','C','D']
sizes=[30,20,10,10]
colors=['gold','yellowgreen','lightpink','pink']
explode=(0.2,0,0,0)

plt.pie(sizes,labels=labels,explode=explode,colors=colors,autopct='%1.1f%%',shadow=True)
```
