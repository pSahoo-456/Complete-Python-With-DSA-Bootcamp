# Inheritnace 
INheritance is a fundamental concept in OOP that allows a class to inherit attributes and methods from another class . 

This lesson covers single inheitance and multiple inheritance . 

```python
#Inheritance Example (Single Inheritance)
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def drive(self):
        print(f"The person is driving the {self.enginetype} car")
```

```python
car1=Car(3,5,"petrol")
car2=Car(4,6,"diesel")

car1.drive()
car2.drive()
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        super().__init__(windows,doors,enginetype)
        self.is_selfdriving=is_selfdriving

    def self_drive(self):
            print(f"Tesla supports self driving : {self.is_selfdriving}")
            
```

```python
tesla1=Tesla(4,4,"electric",True)
tesla1.drive()
tesla1.self_drive()
```

```python
## Multiple Inheritance Example
#Base class 1
class ElectricVehicle:
    def __init__(self, battery_capacity):
        self.battery_capacity = battery_capacity

    def charge(self):
        print(f"Charging the vehicle with battery capacity: {self.battery_capacity} kWh")
#Base class 2
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def info(self):
        print(f"Vehicle Make: {self.make}, Model: {self.model}")
#Derived class
class ElectricCar(Vehicle, ElectricVehicle):
    def __init__(self, make, model, battery_capacity):
        Vehicle.__init__(self, make, model)
        ElectricVehicle.__init__(self, battery_capacity)
        #instead of super() we can also use the above method to call the constructors of both base classes
    def display_info(self):
        self.info()
        print(f"Battery Capacity: {self.battery_capacity} kWh")
```

```python
ecar=ElectricCar("Tesla", "Model S", 100)
ecar.display_info()
ecar.charge()
ecar.info()
```
