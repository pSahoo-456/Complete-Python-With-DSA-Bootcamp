# Polymorphism

Polymorphism is a core concept in OOP. It provides a way to perform a single action in different forms. polymorphism is typically achieved through method overriding and interfaces

## Method Overriding 
Method overriding allows a child class to provide a specific implementation of a method that is already defined in its parent class.

```python
#Base class
class Animal:
    def speak(self):
        return "Animal speaks"
## Derieved class1
class Dog(Animal):
    def speak(self):
        return "Woof!"
## Derieved class2
class Cat(Animal):
    def speak(self):
        return "Meow!"

dog=Dog()
cat=Cat()
print(dog.speak())  # Output: Woof!
print(cat.speak())  # Output: Meow!
```

```python
### Ploymorphism with functions and methods
class Shape:
    def area(self):
        return "The area of the shape"
    
    #Derieved class1
class Circle(Shape):
    def area(self):
        return "The area of the circle is πr²"
    
    #Derieved class2    
class Square(Shape):
    def area(self):
        return "The area of the square is a²"
shapes = [Circle(), Square()]
for shape in shapes:
    print(shape.area()) # Calls the appropriate area method based on the object type    


```
